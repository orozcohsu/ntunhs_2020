```python
import seaborn as sns
import pandas as pd
```

```python
sql="SELECT a.emp_no, salary, birth_date, gender FROM \
       (select distinct emp_no from ntuhs.dept_emp) a\
    join \
       (select emp_no, max(salary) as salary \
        from ntuhs.salaries group by emp_no) b\
    join ntuhs.employees c\
    on a.emp_no = b.emp_no and a.emp_no = c.emp_no"
```

```python
import pymysql
db = pymysql.connect("mysql","root","1234","ntuhs" )
cursor = db.cursor()
cursor.execute(sql)
```

```python
df=pd.DataFrame(cursor.fetchall())
df = df.rename(columns={0: "emp_no", 1: "salary", 2: "birth_date", 3:"gender"})
```

```python
df.head()
```

```python
sns.countplot(df['gender'])
```

```python
#kde 

sns.kdeplot(df['salary'])
```

```python
#kde

ax = sns.kdeplot(df.query('salary < 60000').salary)
ax.set_title("Employee Salary")
```

```python
#histogram

ax = sns.distplot(df['salary'], bins = 20, kde = False)
```

```python
#boxplot

df1=df[['salary']]
sns.boxplot(
    x = 'salary',
    data = df1
)
```

```python
# Multivariant 

df['birth_year']=df['birth_date'].apply(lambda x: int(str(x).split('-')[0]))
g = sns.lmplot(x='birth_year', 
           y='salary',
           hue = 'gender',
           data = df)
g.set_xticklabels(rotation=90)
```

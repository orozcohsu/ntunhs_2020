```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#df=df.set_index('title')
ax = df['count'].plot.bar(
    figsize = (12,6),
    fontsize = 20
)
ax.set_title("Title count", fontsize = 20)
ax.set_xlabel("Title", fontsize = 18)
ax.set_ylabel("Count", fontsize = 18)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('salary_bins.csv')
df
```

```python
df.shape
```

```python
#bar plot 
df['dept_name'].value_counts().plot.bar()
```

```python
#bar plot 
df['bins'].value_counts().plot.bar()
```

```python
df['prefix_bin']=df['bins'].apply(lambda x: int(x.split('_')[0]))
df.head(5)
```

```python
df_p=df[df['dept_name']=='Production'].sort_values(by=['prefix_bin'])
df_s=df[df['dept_name']=='Sales'].sort_values(by=['prefix_bin'])
lst=set(df_p['prefix_bin']).union(set(df_s['prefix_bin']))
lst=sorted(lst)
lst
```

```python
p=[]
s=[]
bin_name=[]

for _ in lst:
    bin_name.append(_)
    try:
        value_p=df_p[df_p['prefix_bin']==_]['count'].values[0]    
    except:
        value_p=0
    p.append(value_p)
    
for _ in lst:
    try:
        value_s=df_s[df_s['prefix_bin']==_]['count'].values[0]
    except:
        value_s=0
    s.append(value_s) 
```

```python
df_ps = pd.DataFrame({'Production': p, 'Sales': s}, index=bin_name)
df_ps.plot.bar(rot=0)
```

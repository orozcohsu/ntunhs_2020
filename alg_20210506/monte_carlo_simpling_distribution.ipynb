```python
from numpy.random import normal
from numpy.random import uniform
from numpy.random import exponential
from numpy.random import laplace

from matplotlib import pyplot
```

```python
# define the distribution
mu = 50
sigma = 5
scale = 2
loc = 0

# generate monte carlo samples of differing size
sizes = [10, 50, 100, 1000]
```

```python
for i in range(len(sizes)):
    # generate sample
    sample = normal(mu, sigma, sizes[i])
    # plot histogram of sample
    pyplot.subplot(2, 2, i+1)
    pyplot.hist(sample, bins=20)
    pyplot.title('%d samples' % sizes[i])
    pyplot.xticks([])
# show the plot
pyplot.show()
```

```python
for i in range(len(sizes)):
    # generate sample
    sample = uniform(mu, sigma, sizes[i])
    # plot histogram of sample
    pyplot.subplot(2, 2, i+1)
    pyplot.hist(sample, bins=20)
    pyplot.title('%d samples' % sizes[i])
    pyplot.xticks([])
# show the plot
pyplot.show()
```

```python
for i in range(len(sizes)):
    # generate sample
    sample = exponential(scale, sizes[i])
    # plot histogram of sample
    pyplot.subplot(2, 2, i+1)
    pyplot.hist(sample, bins=20)
    pyplot.title('%d samples' % sizes[i])
    pyplot.xticks([])
# show the plot
pyplot.show()
```

```python
for i in range(len(sizes)):
    # generate sample
    sample = laplace(loc, scale, sizes[i])
    # plot histogram of sample
    pyplot.subplot(2, 2, i+1)
    pyplot.hist(sample, bins=20)
    pyplot.title('%d samples' % sizes[i])
    pyplot.xticks([])
# show the plot
pyplot.show()
```

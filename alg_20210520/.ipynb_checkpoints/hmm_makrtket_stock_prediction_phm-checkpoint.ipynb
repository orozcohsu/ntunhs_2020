```python
import numpy as np 
from hmmlearn import hmm
import pandas as pd

import os
import tarfile
import pandas as pd
from pandas import errors as pd_errors
from functools import reduce
import logging
import datetime
import sklearn.metrics as metrics
```

```python
# load data
df = pd.read_csv('phm.csv')

#設定日期為index
df['Date']=df['Date'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y").strftime("%Y-%m-%d"))

#日期由小到大排列
df=df.reindex(index=df.index[::-1])

#df['Date'] = pd.to_datetime(df['Date'])
#df=df.set_index('Date', drop=True)

#建立特徵變數
df = df.assign(High5=df.set_index(pd.to_datetime(df.Date)).High.rolling('5d').max().values)
df = df.assign(Low5=df.set_index(pd.to_datetime(df.Date)).Low.rolling('5d').max().values)
df = df.assign(Open5=df.set_index(pd.to_datetime(df.Date)).Open.rolling('5d').max().values)
df = df.assign(Volume5avg=df.set_index(pd.to_datetime(df.Date)).Volume.rolling('5d').mean().values/1000000)
#存檔給模型讀取
df.to_csv('phm_new.csv',index=False)
```

```python
import warnings
import logging
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Supress warning in hmmlearn
warnings.filterwarnings("ignore")

plt.style.use('ggplot')

test_size=0.15
n_hidden_states = 4

class StockPredictor(object):
    def __init__(self, test_size=test_size,
                 n_hidden_states=n_hidden_states, 
                 n_latency_days=10,
                 n_steps_frac_change=50, 
                 n_steps_frac_high=10,
                 n_steps_frac_low=10,
                 #新特徵
                 n_steps_frac_change1=10,
                 n_steps_frac_volume=10):
        
        self._init_logger() 
        self.n_latency_days = n_latency_days
        self.hmm = GaussianHMM(n_components=n_hidden_states)
        self._split_train_test_data(test_size)
        self._compute_all_possible_outcomes(
            n_steps_frac_change, 
            n_steps_frac_high, 
            n_steps_frac_low, 
            #新特徵
            n_steps_frac_change1,
            n_steps_frac_volume)
 
    def _init_logger(self):
        self._logger = logging.getLogger(__name__)
        handler = logging.StreamHandler()
        formatter = logging.Formatter(
            '%(asctime)s %(name)-12s %(levelname)-8s %(message)s')
        handler.setFormatter(formatter)
        self._logger.addHandler(handler)
        self._logger.setLevel(logging.DEBUG)
 
    def _split_train_test_data(self, test_size):
        #讀取新檔案
        data = pd.read_csv('phm_new.csv')
        #shuffle=False(不打亂樣本順序)
        _train_data, test_data = train_test_split(
            data, test_size=test_size, shuffle=False)
        print("total:",len(data))
        print("min train:",min(_train_data['Date']))
        print("max train:",max(_train_data['Date']))
        print("min test:",min(test_data['Date']))
        print("min test:",max(test_data['Date']))
        self._train_data = _train_data
        self._test_data = test_data
 
    @staticmethod
    def _extract_features(data):
        open_price = np.array(data['Open'])
        close_price = np.array(data['Close/Last'])
        high_price = np.array(data['High'])
        low_price = np.array(data['Low'])
        #新特徵
        High5 = np.array(data['High5'])
        Low5 = np.array(data['Low5'])
        Open5 = np.array(data['Open5'])
        Volume = np.array(data['Volume'])
 
        # 計算收盤價、高價和低價的分數變化
        # 這會用到一個特徵
        frac_change = (close_price - open_price) / open_price
        frac_high = (high_price - open_price) / open_price
        frac_low = (open_price - low_price) / open_price
        
        #新特徵
        frac_change1 = (High5 - Low5) / Open5
        frac_volume = (close_price*Volume)
 
        #新特徵
        return np.column_stack((frac_change, frac_high, frac_low, frac_change1, frac_volume))
 
    def fit(self):
        self._logger.info('>>> Extracting Features')
        feature_vector = StockPredictor._extract_features(self._train_data)
        self._logger.info('Features extraction Completed <<<')
 
        self.hmm.fit(feature_vector)
 
    def _compute_all_possible_outcomes(self, 
                                       n_steps_frac_change,
                                       n_steps_frac_high, 
                                       n_steps_frac_low,
                                       #新特徵
                                       n_steps_frac_change1,
                                       n_steps_frac_volume):
        frac_change_range = np.linspace(-0.1, 0.1, n_steps_frac_change)
        frac_high_range = np.linspace(0, 0.1, n_steps_frac_high)
        frac_low_range = np.linspace(0, 0.1, n_steps_frac_low)
        #新特徵
        frac_change_range1 = np.linspace(0, 0.1, n_steps_frac_change1)
        frac_change_volume = np.linspace(0, 0.1, n_steps_frac_volume)
 
        self._possible_outcomes = np.array(list(itertools.product(
            frac_change_range, 
            frac_high_range, 
            frac_low_range, 
            #新特徵
            frac_change_range1,
            frac_change_volume)))
        print("所有的組合的次數:",self._possible_outcomes.shape)
 
    def _get_most_probable_outcome(self, day_index):
        #選擇觀察的資料天數(起迄)
        previous_data_start_index = max(0, day_index - self.n_latency_days)
        previous_data_end_index = max(0, day_index - 1)
        print("previous_data_start_index:",previous_data_start_index)
        print("previous_data_end_index:",previous_data_end_index)
        #抓出起迄範圍的資料集
        previous_data = self._test_data.iloc[previous_data_start_index:previous_data_end_index]
        #將該資料集放入特徵抓取
        previous_data_features = StockPredictor._extract_features(previous_data)
        print("output:",len(previous_data_features))
        
        outcome_score = []
        #計算所有組合
        for possible_outcome in self._possible_outcomes:
            print("previous_data_features:",previous_data_features)
            total_data = np.row_stack((previous_data_features, possible_outcome))
            outcome_score.append(self.hmm.score(total_data))
        most_probable_outcome = self._possible_outcomes[np.argmax(
            outcome_score)]
 
        return most_probable_outcome
 
    def predict_close_price(self, day_index):
        print("day_index:",day_index)
        open_price = self._test_data.iloc[day_index]['Open']
        #增加接收變數
        predicted_frac_change, _, _, _, _ = self._get_most_probable_outcome(day_index)
        
        return open_price * (1 + predicted_frac_change)
 
    def predict_close_prices_for_days(self, days, with_plot=False):
        predicted_close_prices = []
        for day_index in tqdm(range(days)):
            predicted_close_prices.append(self.predict_close_price(day_index))
        if with_plot:
            test_data = self._test_data[0: days]
            #days = np.array(test_data['Date'], dtype="datetime64[ms]")
            days = np.array(test_data['Date'])
            actual_close_prices = test_data['Close/Last']
 
            fig = plt.figure(figsize=(12, 8), dpi=120)
 
            axes = fig.add_subplot(111)
            axes.plot(days, actual_close_prices, 'bo-', label="actual")
            axes.plot(days, predicted_close_prices, 'r+-', label="predicted")
            axes.set_title('PHM')
 
            fig.autofmt_xdate()
            plt.savefig('plot.png') 

            plt.legend()
            plt.show()
            
        
        
            self.predicted_close_prices = predicted_close_prices
 
        return predicted_close_prices

    def diff(self):
        from datetime import datetime
        #date_format = "%Y-%m-%d"
        #a = datetime.strptime(min(self._test_data['Date']), date_format)
        #b = datetime.strptime(max(self._test_data['Date']), date_format)
        #delta = b - a
        print("測試資料集天數:",len(self._test_data))


    def evaluation(self, days):
        print("MAE:", metrics.mean_absolute_error(self._test_data['Close/Last'][0:days], self.predicted_close_prices))
        print("MSE:", metrics.mean_squared_error(self._test_data['Close/Last'][0:days], self.predicted_close_prices))
        print("MedAE:", metrics.median_absolute_error(self._test_data['Close/Last'][0:days], self.predicted_close_prices))
        print("RSQ:", metrics.r2_score(self._test_data['Close/Last'][0:days], self.predicted_close_prices))
```

```python
stock_predictor = StockPredictor()
stock_predictor.fit()
stock_predictor.diff()
```

```python
#prediction days
days=6

stock_predictor = StockPredictor()
stock_predictor.fit()
res=stock_predictor.predict_close_prices_for_days(days, with_plot=True)
```

```python
stock_predictor.evaluation(days)
```
